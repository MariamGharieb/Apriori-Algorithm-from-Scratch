Apriori Algorithm 

```python
from collections import Counter
c = Counter() #c is list of integers -each integer is the frequent of item in data
#calculate the frequent of each item
for i in init:
    for d in data:
        if(i in d[1:4]): #mmkn yelf 3la eldata in 3 col
            c[i]+=1
#just printing the counters
print("Counting list #1 :")
for i in c:
    print(str([i])+": "+str(c[i]))
print()

l = Counter()
#met the mini support requirement
for i in c:
    if(c[i] >= s):
        l[frozenset([i])]+=c[i]
#printing the list after mini support conidtion

print("List :")
for i in l:
    print(str(list(i))+": "+str(l[i]))
print()

pl = l
pos = 1 
#add new item
for count in range (2,5):
    nc = set()
    temp = list(l)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
            t = temp[i].union(temp[j])
            if(len(t) == count):
                nc.add(temp[i].union(temp[j]))
    nc = list(nc)
    #counting new items 0
    c = Counter()
    for i in nc:
        c[i] = 0
        for q in data:
            temp = set(q[1:4]) # temp = data
            if(i.issubset(temp)): # compute the freq of the new union
                c[i]+=1
            #printing new items iterations
    print("Counting list #"+str(count)+":")
    for i in c:
        print(str(list(i))+": "+str(c[i]))
    print()
    l = Counter()
    for i in c:
        if(c[i] >= s):
            l[i]+=c[i]
    
    print("List :"+str(count)+":")
    for i in l:
        print(str(list(i))+": "+str(l[i]))
    print()
    if(len(l) == 0):
        break
    pl = l
    pos = count
print("Result: ")
print("List :"+str(pos)+":")
for i in pl:
    print(str(list(i))+": "+str(pl[i]))
print()
```

#Association Rule

```python
from itertools import combinations
mmax = conf
#print("Association Rules:")
for l in pl:
    c = [frozenset(q) for q in combinations(l,len(l)-1)] 
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q[1:4])
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        #print (str(list(a))+" -> "+str(list(b))+" = "+str(sab/sa*100)+"%")
        temp = sab/sa*100
        if(temp >= mmax):
            print( "Result = " + str(list(a))+" -> "+str(list(b))+" = "+str(sab/sa*100)+"%")
            
        temp = sab/sb*100
        if(temp >= mmax):
            print("Result = " +  str(list(b))+" -> "+str(list(a))+" = "+str(sab/sb*100)+"%")
            
```
